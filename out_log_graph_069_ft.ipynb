```python
import matplotlib.pyplot as plt

f = open( 'out_069_ft.log', mode = 'r', encoding='utf-8')

line = f.readline()

i = 0
i2 = 0
x = []
x2 = []
loss = []
error = []
lr = []
v_i = 0
v_i2 = 0
v_x = []
v_x2 = []
v_loss = []
v_lr = []
v_error = []
while line:
    line = f.readline()
    if 'phase:    train,' in line :
        i += 1
        x.append( i * 5 / 585)
        line_split = line.split(',')
        loss.append( float(line_split[2].split(':')[1]) )
        lr.append( float(line_split[3].split(':')[1]) )
    if 'phase:validation,' in line :
        v_i += 1
        v_x.append( v_i * 5 / 25 )
        line_split = line.split(',')
        v_loss.append( float(line_split[2].split(':')[1]) )
        v_lr.append( float(line_split[3].split(':')[1]) )
    if '    train token error rate:' in line:
        #print( "line", line)
        i2 += 1
        x2.append( i2 )
        line_split = line.split( ' ' )
        #print( line_split)
        error.append( float( line_split[8]))
    if 'validation token error rate:' in line:
        v_i2 += 1
        v_x2.append( v_i2 )
        line_split = line.split( ' ' )
        #print( line_split)
        v_error.append( float( line_split[8]))

plt.plot( x, loss, label='Train Loss')
plt.plot( v_x, v_loss, label='Validation Loss' )
plt.title( "Loss")
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss')
plt.legend()

plt.show()
```

```python
#plt.plot( x2, error, label = 'Train Token Error')
plt.plot( v_x2, v_error, label='Validation Token Error')
plt.title( "Token Error")
plt.xlabel( 'Epochs')
plt.ylabel( 'Error %')
plt.legend()

print( v_error )
plt.show()
```
