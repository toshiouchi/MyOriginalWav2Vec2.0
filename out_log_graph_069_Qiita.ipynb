```python
import matplotlib.pyplot as plt

f = open( 'out_069.log')

line = f.readline()

i = 0
x = []
lm = []
ld = []
loss = []
pos_avg = []
neg_avg = []
lr = []
i2 = 0
v_x = []
v_lm = []
v_ld = []
v_loss = []
v_pos_avg = []
v_neg_avg = []
while line:
    line = f.readline()
    if 'phase:    train,' in line :
        i += 1
        x.append( i * 5 /4750)
        line_split = line.split(',')
        lm.append( float(line_split[2].split(':')[1]) )
        #print( line )
        ld.append( float(line_split[3].split(':')[1]) )
        loss.append( float(line_split[4].split(':')[1]) )
        pos_avg.append( float(line_split[5].split(':')[1]) )
        neg_avg.append( float(line_split[6].split(':')[1]) )
        lr.append( float(line_split[7].split(':')[1]) )
    if 'phase:validation,' in line :
        i2 += 1
        v_x.append( i2 * 5 / 125 )
        line_split = line.split(',')
        v_lm.append( float(line_split[2].split(':')[1]) )
        v_ld.append( float(line_split[3].split(':')[1]) )
        v_loss.append( float(line_split[4].split(':')[1]) )
        v_pos_avg.append( float(line_split[5].split(':')[1]) )
        v_neg_avg.append( float(line_split[6].split(':')[1]) )

plt.plot( x, loss, label='train')
plt.plot( v_x, v_loss, label= 'validation' )
plt.title( "Loss")
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss')
plt.legend()

plt.show()
```

```python
plt.plot( x, pos_avg, label='train positive average' )
plt.plot( x, neg_avg, label='train negative average' )
plt.plot( v_x, v_pos_avg, label = "val positive average")
plt.plot( v_x, v_neg_avg, label="val negative average")

plt.title( "Similarity")
plt.xlabel( "Epochs")
plt.ylabel( "Similarity")
plt.legend()

plt.show()
```

```python
plt.plot( x, lm , label="train Lm")
plt.plot( v_x, v_lm, label="validation Lm")
plt.title( "Lm")
plt.xlabel( "Epochs")
plt.ylabel( "Lm")
plt.legend()

plt.show()
```

```python
plt.plot( x, ld, label="train Ld" )
plt.plot( v_x, v_ld, label="validation Ld")
plt.title( "Ld")
plt.xlabel( "Epochs")
plt.ylabel( "Ld")
plt.legend()
plt.show()
```

```python
plt.plot( x, lr)
plt.title( "Learning Rate")
plt.xlabel( "Epochs")
plt.ylabel( "Lr")
plt.show()
```

```python
plt.plot( v_x, v_loss)

plt.title( "Validation Loss")
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss')

plt.show()
```
